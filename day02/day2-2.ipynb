✅ 1. for 반복문

정해진 횟수만큼 반복할 때 사용합니다.

리스트, 문자열, 범위 등 반복 가능한 객체를 순회합니다.

```python
for num in range(1,4):    
    print(f'{num}. 안녕하세요')
```

```python
for i in range(1,10):  # 1 ~ 9  
    print(f'2 X {i} = {2*i}')
```

```python
for i in range(1,6):
    print('*' * i)
```

```python
sample1 = [1,5,7,9,10,11,12,13,14]
sample2 = []
for i in sample1:
    if i >=10:
        sample2.append(i)
```

```python
sample3 = (1,2,3,4,5,6,7,8,9)
sample4 = []
# [1,5,10,15]
for i in sample3:
    if i in [1,5,10,15]:
        sample4.append(i)

sample4
```

✅ 2. while 반복문

조건이 참(True) 인 동안 계속 반복합니다.

조건을 만족하지 않으면 종료합니다.

```python
count = 0
while count < 5:
    print('안녕하세요')
    count = count + 1
```

```python
for i in range(2,10):
    for j in range(1,10):
        print(f'{i} x {j} = {i * j } ')
    print('-' * 100)
```

```python
import random
random.sample(range(100), 10)  # range(100) 범위중에서 중복되지않은 랜덤한 데이터 10개 추출해서 리스트로 반환
```

```python
# 랜덤한 샘플데이터
samples = random.sample(range(1000), 20)
print(samples)
# 이중에서 짝수만 찾아서 변수에 저장 --> 리스트
# even_lists 변수에 저장
even_lists = []
for i in samples:
    if i % 2 == 0 :  #짝수
        even_lists.append(i)
print(even_lists)        
```
