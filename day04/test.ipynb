```python
import csv
import re

def load_exchange_rates_from_bok_csv(filename="data.csv"):
    """
    CSV íŒŒì¼ì—ì„œ í™˜ìœ¨ ì •ë³´ë¥¼ ì½ê³ , 'í†µí™”ì´ë¦„': í™˜ìœ¨(1 KRW ê¸°ì¤€) ë”•ì…”ë„ˆë¦¬ ë°˜í™˜
    ì˜ˆ: {"ë¯¸êµ­ë‹¬ëŸ¬": 0.000733, "ìœ„ì•ˆ": 0.00528, ...}
    """
    rates = {}

    try:
        with open(filename, newline='', encoding='utf-8-sig') as csvfile:
            reader = list(csv.reader(csvfile))

            header = reader[0]
            latest_date_index = len(header) - 1
            latest_date = header[latest_date_index]

            for row in reader[1:]:
                account = row[1]  # ì˜ˆ: "ì›/ë¯¸êµ­ë‹¬ëŸ¬(ë§¤ë§¤ê¸°ì¤€ìœ¨)"
                rate_str = row[latest_date_index].replace(',', '').strip()

                # "ì›/ë¯¸êµ­ë‹¬ëŸ¬(ë§¤ë§¤ê¸°ì¤€ìœ¨)" â†’ "ë¯¸êµ­ë‹¬ëŸ¬" ì¶”ì¶œ
                match = re.search(r"ì›/([^()]+)", account)
                if not match:
                    continue
                currency_name = match.group(1).strip()

                try:
                    rate = float(rate_str)
                    rates[currency_name] = 1 / rate  # 1 KRW â†’ ì™¸í™”
                except ValueError:
                    print(f"âŒ ì˜ëª»ëœ í™˜ìœ¨ ë°ì´í„°: {account} - '{rate_str}'")
                    continue

        return rates, latest_date

    except FileNotFoundError:
        print(f"âŒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {filename}")
        return {}, None


def convert_currency(amount_krw, rate):
    return amount_krw * rate


def main():
    print("ğŸ’± í•œêµ­ì€í–‰ í™˜ìœ¨ ê¸°ë°˜ ì›í™” í™˜ì „ê¸°")
    rates, rate_date = load_exchange_rates_from_bok_csv("data.csv")

    if not rates:
        print("âŒ í™˜ìœ¨ ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
        return

    print("\nğŸŒ ì‚¬ìš© ê°€ëŠ¥í•œ í†µí™” ëª©ë¡:")
    for name in rates:
        print(f"- {name}")

    target_currency = input("\në³€í™˜í•  ì™¸í™”ë¥¼ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: ë¯¸êµ­ë‹¬ëŸ¬, ìœ„ì•ˆ, ìœ ë¡œ): ").strip()

    if target_currency not in rates:
        print("âŒ ì§€ì›ë˜ì§€ ì•ŠëŠ” í†µí™”ì…ë‹ˆë‹¤.")
        return

    try:
        amount_krw = float(input("ë³€í™˜í•  ê¸ˆì•¡ì„ ì…ë ¥í•˜ì„¸ìš” (KRW): ").strip())
        if amount_krw < 0:
            raise ValueError
    except ValueError:
        print("âŒ ìœ íš¨í•œ ì–‘ìˆ˜ ê¸ˆì•¡ì„ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤.")
        return

    rate = rates[target_currency]
    converted = convert_currency(amount_krw, rate)

    print(f"\nğŸ’¹ ê²°ê³¼:")
    print(f"{amount_krw:,.2f} KRW â†’ {converted:,.2f} {target_currency}")
    print(f"(í™˜ìœ¨: {rate:.6f}, ê¸°ì¤€ì¼: {rate_date})")


if __name__ == "__main__":
    main()
```
